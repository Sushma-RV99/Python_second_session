```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.bikewale.com/new-bike-search/best-scooty-under-1-lakh/")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
soup
```

```python
all=soup.find_all("div",{"class":"o-bQ o-cm o-cA o-c4 o-kY o-f o-aE"})
all
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("span",{"class":"o-jJ o-j3 o-js o-ez"})[0].text)
    print(item.find_all("span",{"class":"o-jJ o-jr o-j3 o-kJ"})[0].text)
    print(item.find_all("p",{"class":"o-j1 o-cx o-jq o-d o-j o-jJ"})[0].text)
    print(item.find_all("div",{"class":"o-o onOEBQ o-iT o-ei"})[0].text)
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["bike_name"]=item.find_all("span",{"class":"o-jJ o-j3 o-js o-ez"})[0].text
   
    d["bike_price"]=item.find_all("span",{"class":"o-jJ o-jr o-j3 o-kJ"})[0].text.replace("","Rs.")
    d["rating"]=item.find_all("p",{"class":"o-j1 o-cx o-jq o-d o-j o-jJ"})[0].text
    d["features"]=item.find_all("div",{"class":"o-o onOEBQ o-iT o-ei"})[0].text
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```
